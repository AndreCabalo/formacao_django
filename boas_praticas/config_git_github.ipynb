# Variáveis de ambiente

Queremos versionar o código, mas não podemos versionar diretamente, pois temos dados restritos no app

E provavelmente o proprio github nos retornaria um mensagem que estamos nos expondo nossa chave secreta Django (linha 23) do arquivo settings

## Para evitar de expor alguns conteúdo 

Podemos:

- Criar as variaveis de ambiente, para isso existe um pacote chamado python-dotenv, da seguinte maneira:

<code>pip install python-dotenv</code>

- Não esquece de atualizar o requirements.txt

- Na sequencia, criaremos um arquivo (fora da pasta "setup" chamado de ".env"

- E criaremos uma variavel chamada SECRET_KEY, igual na variavel do arquivo settingS e colaremos dentro da nova variavel com mesmo nome, porém sem as aspas 

- No arquivo Settings, limparemos o conteudo da variavel

- Apontaremos que teremos variaveis de ambientes para serem carregadas cno topo do arquivo settings:



```python
from pathlib import Path, os
from dotenv import load_dotenv

load_dotenv()
```

E onde ficava a SECRET_KEY no arquivo setting, iremos inserir o seguintw:

```python
# SECURITY WARNING: keep the secret key used in production secret!
SECRET_KEY = str(os.getenv('SECRET_KEY'))
```

Com isso nosso projeto voltará a funcionar normalmente e desta vez, não expondo os dados, podendo assim ser carregado no git (apenas o .env, que não poderemos enviar)

!!Não enviaremos o arquivo .env para o github!!

# Criando repositório no Github

Criamos um repositório publico no github, chamado de Alura-Space

## Criamos o .gitigonore

E como queremos ignorar alguns conteudo, criamos dentro da pasta alura-space no vs-code o arquivo .gitignore e nesse arquivo colocaremos os arquivos que queremos ignorar

Precisamos inserir tudo que queremos ignorar, e temos um forte aliado para definir isso, o site gitignore.io

## GitIgnore.io

Entrando no site gitignoreio, temos um interface, na qual questiona qual linguagem vc esta codando, e então ele faz um lista de tudo que devemos ignorar, por tanto vamos inserir django e o resultado é:



e aparece um serie de código que devemos dar :

<code>ctrl a + ctrl c </code>

Voltamos no nosso arquivo ".gitignore" e damos <code>ctrl+c</code>, dentro ele



## Inicinado um git

Começamos com o comando no terminal, para iniciar o repositório local:

<code>git init</code>

## Adicionando o conteudo

Adiciona tudo, claro respeitando o gitIgnore

<code>git add .</code>

## Criando o comiit

Criamos o commit

<code>git commit -m "projeto alura space"</code>

## Endereçando o repositório remoto

Vamos ao github e pegamos o endereço:



Para apontar que o respositório local, aponte para o repositório no github

<code>git remote add origin https://github.com/AndreCabalo/alura-space.git</code>

## Dando o push

E depois daremos o push com:

<code>git push origin master</code>

Com isso podemos conferir no github, se for respeitado o gitignore.
